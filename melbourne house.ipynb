```python
#Lower hinge
#Q1
Q1=mel['Car'].quantile(0.25)
```

```python
#upper hinge
#Q3
Q3=mel['Car'].quantile(0.75)
```

```python
#interquartile Range
IQR=Q3-Q1
IQR
```

```python
#Lower whisker
FLOOR = Q1 - 1.5 * IQR
FLOOR
```

```python
#upper Whisker
CAP = Q3 + 1.5 * IQR
CAP
```

```python
mel_ist=mel[(mel['Car'] >= FLOOR) &(mel['Car'] <= CAP)]
```
